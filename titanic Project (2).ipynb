```python
# 1. How many passengers survived and how many did not?
survival_counts = titanic["Survived"].value_counts()
#.value_counts()counts how many times each unique value appears in a column
#Counts how many passengers survived (1) and didn’t survive (0)
survival_counts 
```

```python
# 2.Did more women survive than men?
gender_survival = titanic.groupby('Sex')['Survived'].sum() # groups the data by gender and sums the Survived column.
gender_survival
```

```python
#gender_total = titanic.groupby('Sex').sum()
#gender_total
sns.countplot(data=titanic, x='Sex', hue='Survived',palette='viridis')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
## 3.Which class (1st, 2nd, or 3rd) had the most passengers

class_counts = titanic['Pclass'].value_counts() #Counts how many passengers were in each passenger class (1st, 2nd, 3rd).
class_counts
```

```python
## 4.What was the average age of passengers in each class?

avg_age_class = titanic.groupby('Pclass')['Age'].mean()
avg_age_class
avg_age_class.round(0) # round makes the if its 34.84747 to 34.0 if its is 34.84747 to 34.84 round(2)means we can custmize
```

```python
# 5. How many children (under 16) were on board?

children_16 = titanic[titanic["Age"] < 16].value_counts().sum()
children_16
```

```python
## 6.Which port did most passengers board from?

embarked_counts = titanic["Embarked"].value_counts() #value_counts() is used to count how many times each unique value appears in a column.
embarked_counts
```

```python
## 7.How many people traveled alone and how many with family?

# Group passengers by family onboard
travel_group = titanic.groupby((titanic["SibSp"] + titanic["Parch"]) > 0)["PassengerId"].count()
# Adds the number of siblings/spouses (SibSp) and parents/children (Parch) each passenger had on board.
#If result = 0 the passenger is alone
#If result > 0 the passenger is with family
#(titanic["SibSp"] + titanic["Parch"]) > 0(Creates a Boolean series:False → AloneTrue → With Family)
# Rename index for clarity
travel_group.index = ["Alone", "With Family"]
plt.figure(figsize=(5,5))
plt.pie(travel_group, labels=travel_group.index, autopct="%1.0f%%",colors=['red','green'])
plt.title("Traveling Alone vs With Family")
plt.show()

```

```python
##8. What is the average fare for each passenger class?
plt.figure(figsize=(6,4))
sns.barplot(x="Pclass", y="Fare", data=titanic)
plt.title("Average Fare by Class")
plt.show()
```

```python
## 9. Were younger passengers more likely to survive?

plt.figure(figsize=(8,5))
sns.histplot(
    data=titanic,
    x="Age",
    hue="Survived",         
    multiple="stack",       
    bins=20,
    palette="husl",
    edgecolor="black"
)

plt.title("Age Distribution by Survival Status")
plt.xlabel("Age")
plt.ylabel("Passenger Count")
plt.grid()
plt.show()
```
